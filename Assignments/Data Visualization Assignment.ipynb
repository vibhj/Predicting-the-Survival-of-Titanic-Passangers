## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
Stores=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Items=['Clothing','Equipment','Accessories']
Clothing = np.array([82,78,48,46,39])
Equipment = np.array([48,31,51,29,18])
Accessories = np.array([15,20,14,13,9])
var=[x for x,_ in enumerate(Stores)]
plt.bar(var, Clothing, width=0.8, label='Clothing', color='r', bottom=Equipment+Accessories)
plt.bar(var, Equipment, width=0.8, label='Equipment', color='lightgreen', bottom=Accessories)
plt.bar(var, Accessories, width=0.8, label='Accessories', color='b')
plt.xlabel('Stores',size=20,color='m')
plt.xticks(var,Stores)
plt.ylabel('Items in Store',size=20,color='orange')
plt.legend(loc='upper right')
plt.xticks(rotation='90')
plt.title('Stacked bar graph',size=20)
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
df=sns.load_dataset('titanic')
df
```

```python
sns.countplot(x = 'class', hue = 'survived', palette = 'Set1', data = df)
plt.title('Passenger status (Survived/Died) against Passenger Class',size=20)
plt.legend(title='Survived',loc='upper left')
plt.xlabel('Passenger Class',size=20)
plt.ylabel('Survived',size=20)
plt.legend(loc='upper left')
plt.show()
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
tip=sns.load_dataset('tips')
tip
```

```python
sns.set_style('darkgrid')
sns.distplot(tip['total_bill'],color='g',vertical=True)
plt.ylabel('Total_bill',size=20,color='c')
plt.xlabel('Average of Total_bill',size=20,color='c')
plt.title('Horizontal Distribution Graph',size=20,color='c')
plt.show()
```

## Q.9 


- Plot the Pie Chart for the two colums with some explode for the country "India"  and proper shadow

```python
Countries=['United States','Great Britain','India','Russia','Germany']
Medal=[46,27,26,19,17]
color=['c','yellowgreen','lightcoral','lightskyblue','pink']
explode=(0,0,0.4,0,0)
plt.pie(Medal,labels=Countries,colors=color,autopct='%1.1f%%',explode=explode,pctdistance=0.7,shadow=True,startangle=90)
plt.title('Percentage of Medals achieve by the countries',size=20)
plt.show()
```
